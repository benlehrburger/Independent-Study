```python
import os
import selenium
from selenium import webdriver 
import base64
import time
import urllib.request
```

```python
# open browser
DRIVER_PATH = '/usr/local/bin/chromedriver'
SAVE_FOLDER = './scrape'
GOOGLE_IMAGES = 'https://www.google.com/search?q=mid+century+modern+architecture&tbm=isch&ved=2ahUKEwjpy6_MtKD_AhV-MVkFHf-pCcQQ2-cCegQIABAA&oq=mid+cent&gs_lcp=CgNpbWcQARgAMgcIABCKBRBDMgoIABCKBRCxAxBDMgcIABCKBRBDMgcIABCKBRBDMgcIABCKBRBDMgoIABCKBRCxAxBDMggIABCABBCxAzIICAAQgAQQsQMyCggAEIoFELEDEEMyCggAEIoFELEDEEM6BQgAEIAEOgYIABAHEB46BggAEAgQHlD2B1ifDmCdHGgAcAB4AIABUIgB2QSSAQE5mAEAoAEBqgELZ3dzLXdpei1pbWfAAQE&sclient=img&ei=BLB3ZOm5C_7i5NoP_9OmoAw&bih=742&biw=791'
driver = webdriver.Chrome(DRIVER_PATH)
driver.get(GOOGLE_IMAGES)
```

```python
# Scroll to the end of the page
def scroll_to_end():
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    time.sleep(5)
    print('scroll done')
```

```python
# execute scrape

counter = 0
for i in range(1, 10):
    
    scroll_to_end()
    image_elements = driver.find_elements_by_class_name('rg_i')
    
    for image in image_elements: 
        
        if (image.get_attribute('src') is not None):
            
            my_image = image.get_attribute('src').split('data:image/jpeg;base64,')
            filename = SAVE_FOLDER + 'opt'+str(counter)+'.jpeg'
            
            if(len(my_image) >1): 
                
                with open(filename, 'wb') as f: 
                    f.write(base64.b64decode(my_image[1]))
                    
            else: 
                print(image.get_attribute('src'))
                urllib.request.urlretrieve(image.get_attribute('src'), SAVE_FOLDER + 'helmet'+ str(counter)+'.jpeg')
            
            counter += 1
```
